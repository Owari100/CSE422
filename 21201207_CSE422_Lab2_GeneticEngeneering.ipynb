```python
# Genetic Algorithm

import random
import math

inp_f = open("Input_1.txt")
out_f = open("Output_1.txt", "w")

list_inp = (inp_f.readlines())
lenth_class_time = (list_inp[0].strip().split(' '))
lenth_time = int(lenth_class_time[0])
lenth_class = int(lenth_class_time[1])

popu = []
if lenth_class < lenth_time or lenth_class == lenth_time:
      for i in range(0, 150, 1):
          chro = ''
          for s in range(0, lenth_time*lenth_class, 1):
              chro += str(random.randint(0, 1))
          popu.append(chro)

def penalty_overlap(val_s):
    pen = 0
    for i in range(0, int(len(val_s)/ lenth_time), 1):
            flag = 0
            for j in range(i, len(val_s),3):
                if val_s[j] == '1':
                    flag += 1
            if flag != 0:
                pen = pen + (flag - 1)
    return (pen)

def penalty_consistency(val_s):
    pen = 0
    for i in range(0, len(val_s), 3):
        flag = 0
        for j in range(i, i+3, 1):
                if val_s[j] == '1':
                    flag += 1
        if flag != 0:
            pen = pen + (flag - 1)
    return (pen)

store = {}
def fit_function(val_s):
    fit_pen = -(penalty_overlap(val_s) + penalty_consistency(val_s))
    store[val_s] = fit_pen
    return (fit_pen)

def crossover(par_1, par_2):
    mid = int(len(par_1) / 2)
    val_1 = par_1[:mid] + par_2[mid:]
    val_2 = par_1[mid:] + par_2[:mid]
    popu.append(val_1)
    popu.append(val_2)
    return (val_1, val_2)

def mutation(val_s):
    ind = int(str(random.randint(0, len(val_s)-1)))
    if val_s[ind] == '0':
        mut = val_s[:ind] + '1' + val_s[ind+1:]
    else:
        mut = val_s[:ind] + '0' + val_s[ind+1:]
    return (mut)

def genetic_algorithm(par_1, par_2, boundary):
    val_x, val_y = crossover(par_1, par_2)
    if fit_function(val_x) == 0:
        return (val_x)
    elif fit_function(val_y) == 0:
        return (val_y)
    else:
        boundary -= 1
        if boundary == 0:
            return ("000000000")
        val_x = mutation(val_x)
        val_y = mutation(val_y)
        return (genetic_algorithm(val_x, val_y, boundary))

x =  popu[(random.randint(0, len(popu)))]
y =  popu[(random.randint(0, len(popu)))]
ans = genetic_algorithm(x, y, 1000)
if ans == "000000000" or ans == "111111111":
  print('-100', file = out_f)
else:
    print(ans, file = out_f)
    print(store[ans], file = out_f)

inp_f.close()
out_f.close()
```

```python
from random import randint as test
import math

inn = open("Input_1.txt")
inn = inn.readlines()
size = inn[0].strip().split()
size = int(size[0])
print(size)
l = []

for i in inn[1:]:
  listt = i.split()
  print(listt)
  if listt[0] == "l":
    l.append("-"+listt[0])
  else:
    l.append(listt[0])
int_l = [int(i) for i in l ]
def fit_func(p):
  fit_score = 0
  for i in range(len(p)):
    if p[i] == "1":
      fit_score += int_l[i]

  return fit_score
def crossover(parent1 , parent2):
    midpoint = int(math.ceil(size/2.0))
    x = parent1[:midpoint]+parent2[midpoint:]
    y = parent2[:midpoint]+parent1[midpoint:]

    return x,y
def mutation(parent):
    index = int(str(test(0, size-1)))
    if parent[index]=="1":
        change = "0"
    else:
        change = "1"
    mutant = parent[:index]+change+parent[index+1:]
    return mutant
def randPopulation(size):
    string = ""
    for i in range(size):
        string+=str(test(0, 1))

    return(string)
def geneticAlgo(parent1, parent2, lim):
    x, y = crossover(parent1,parent2)
    if (fit_func(x)==0):
      return x
    elif (fit_func(y)==0):
      return y
    else:
      lim-=1
      x = mutation(x)
      y = mutation(y)
      return geneticAlgo(x,y,lim)
x = randPopulation(size)

y = randPopulation(size)

result = geneticAlgo(x, y, 1000 )
if result == "00000":
  print(-1)
else:
  print(result)
```
